```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge
```

```python
df=pd.read_csv('student_scores.csv')
```

```python
df.head(3)
```

```python
y=df[['Scores']]
X=df[['Hours']]
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8,8))
plt.scatter(X,y)
plt.show()
```

```python
lr=LinearRegression()
model=lr.fit(X,y)
model.score(X,y)
```

```python
alfalar=[1,10,20,100,200,0.1]
for a in alfalar:
    r=Ridge(alpha=a)
    modelr=r.fit(X,y)
    skor=modelr.score(X,y)
    print("Skor",skor)
    print("Katsayı",modelr.coef_)
```
